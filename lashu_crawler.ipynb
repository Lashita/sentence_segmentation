```python
import requests #to make url requests
from bs4 import BeautifulSoup #To parse web files
```

```python
def night_crawler(webpage):
    url = str(webpage)
    source_code = requests.get(url) #built in function in requests to pull source code
    plain_text = source_code.text #converting the source code to plain text using .text method
    soup = BeautifulSoup(plain_text,"html.parser") #Feeding this text to BeautifulSoup for analysing
    for link in soup.findAll("a"): #We are trying to find all the links
        if link.get('href') == None:
            continue
        else:
            href = link.get("href") #saving the link in the variable href
            print(href) #printing the link
```

```python
web_page = input("Enter the URL to crawl: ")

night_crawler(web_page)
```

```python
def lashu_crawler(webpage):
    url = str(webpage)
    source_code = requests.get(url)
    plain_text = source_code.text
    soup = BeautifulSoup(plain_text, "html.parser")
    
    #title from meta tags
    title = soup.title.string
    print("The site title is: " + title)
        
    #description from meta tags
    metas = soup.find_all('meta')
    print("The site description is: ")
    print([meta.attrs['content'] for meta in metas if 'name' in meta.attrs and meta.attrs['name'] == 'description'])
    #we are using list comprehensions
        
    #links from page
    print("These are the links: ")
    for link in soup.findAll("a"): 
        if link.get('href') == None:
            continue
        else:
            href = str(webpage) + '/' + link.get("href")
            print(href)
```

```python
web_page = input("Enter the URL to crawl: ")

lashu_crawler(web_page)
```
